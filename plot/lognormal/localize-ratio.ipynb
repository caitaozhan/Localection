```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
import time
import random
plt.rcParams['font.size'] = 25
filename = '{}-{}-{}.txt'
```

```python
num_intru = [1, 3, 5, 7, 10]

mean1 = []
mean2 = []
mean3 = []
mean4 = []

for num in num_intru:
    ratio = np.loadtxt(filename.format(num, 'ours', 'ratio'))
    mean1.append(ratio[0])
    mean2.append(ratio[1])
    mean3.append(ratio[2])
    mean4.append(ratio[3])

mean1 = np.array(mean1)
mean2 = np.array(mean2)
mean3 = np.array(mean3)
mean4 = np.array(mean4)
```

```python
plt.figure(figsize=(12, 9))

ind   = np.arange(len(num_intru))    # the x locations for the groups
width = 0.5
p1 = plt.bar(ind, mean1, width, edgecolor='black', label='Proc-1')
p2 = plt.bar(ind, mean2, width, edgecolor='black', bottom=mean1, label='Proc-1.1')
p3 = plt.bar(ind, mean3, width, edgecolor='black', bottom=mean1+mean2, label='Proc-2,t=2')
p4 = plt.bar(ind, mean4, width, edgecolor='black', bottom=mean1+mean2+mean3, label='Proc-2,t=3')

plt.xticks(ind, ('1', '3', '5', '7', '10'))
plt.xlabel('Varing # of Intruders')

plt.legend()
plt.ylabel('Ratio')
plt.yticks(np.arange(0, 1.1, 0.1))
plt.ylim([0, 1.1])
plt.title('Ratio of Intruder Identified \n by Different Procedures and # of Intruders')
plt.savefig('varyIntruder-ratio.png')
```
