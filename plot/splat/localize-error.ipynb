```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.size'] = 25
```

```python
mean1   = []
std1    = []
mean2   = []
std2    = []
mean3   = []
std3    = []
filename = '{}-{}-{}.txt'
```

```python
num_intru = [1, 3, 5, 7, 10]

for num in num_intru:
    error = np.loadtxt(filename.format(num, 'ours', 'error'), delimiter=',')
    mean1.append(error.mean())
    std1.append(error.std())
    
for num in num_intru:
    error = np.loadtxt(filename.format(num, 'splot', 'error'), delimiter=',')
    mean2.append(error.mean())
    std2.append(error.std())

for num in num_intru:
    error = np.loadtxt(filename.format(num, 'cluster', 'error'), delimiter=',')
    mean3.append(error.mean())
    std3.append(error.std())
```

```python
cell_len = 100
mean1 = np.array(mean1) * cell_len
mean2 = np.array(mean2) * cell_len
mean3 = np.array(mean3) * cell_len
```

```python
ind   = np.arange(len(mean1))
width = 0.24
```

```python
fig, ax = plt.subplots(figsize=(12, 9))
ax.bar(ind-width-0.02, mean1, width, edgecolor='black', label='Our')
ax.bar(ind,            mean2, width, edgecolor='black', label='SPLOT')
ax.bar(ind+width+0.02, mean3, width, edgecolor='black', label='cluster')
ax.set_title('Errors by Algorithms and # of Intruder', fontsize=30)
ax.legend()
ax.set_ylabel('Error (m)')
#ax.set_yticks(fontsize=20)
ax.set_xticks(ind)
ax.set_xticklabels(('1', '3', '5', '7', '10'))
ax.set_xlabel('Varing # of Intruder')
fig.savefig('varyIntruder-error.png')
```
