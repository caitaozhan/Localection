```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.rcParams['font.size'] = 25
```

```python
mean1 = [0.113, 0.386, 0.878]  # 5% training
mean2 = [0.02, 0.137, 0.375]   # 10% training
mean3 = [0.0, 0.098, 0.33]     # 20%
mean4 = [0.0, 0.125, 0.314]    # 50%
mean5 = [0.0, 0.099, 0.263]    # 100%

N     = 3
ind   = np.arange(N)    # the x locations for the groups
width = 0.16       # the width of the bars: can also be len(x) sequence

pos1 = ind - 2*width - 0.02
pos2 = ind - 1*width - 0.01
pos3 = ind - 0*width
pos4 = ind + 1*width + 0.01
pos5 = ind + 2*width + 0.02


fig, ax = plt.subplots(figsize=(20, 12))
ax.bar(pos1, mean1, width, edgecolor='black', label='5%')
ax.bar(pos2, mean2, width, edgecolor='black', label='10%')
ax.bar(pos3, mean3, width, edgecolor='black', label='20%')
ax.bar(pos4, mean4, width, edgecolor='black', label='50%')
ax.bar(pos5, mean5, width, edgecolor='black', label='100%')

ax.set_title('Errors by Percentage of Training Data and # of Intruder')
ax.legend()
ax.set_ylabel('Error (100m)')
#ax.set_yticks(fontsize=20)
ax.set_xticks(ind)
ax.set_xticklabels(('1', '3', '5', '7', '10'))
ax.set_xlabel('Varing # of Intruder')
fig.savefig('varyIntruder-error.png')
```

```python
mean1 = [0.08, 0.23, 0.26]  # 5%  miss
mean2 = [0.0, 0.03, 0.064]  # 5%  false
mean3 = [0.0, 0.025, 0.116]  # 10% miss
mean4 = [0.0, 0.005, 0.032]  # 10% false
mean5 = [0.0, 0.02, 0.084]  # 20% miss
mean6 = [0.0, 0.015, 0.044]  # 20% false
mean7 = [0.0, 0.01, 0.088]  # 50% miss
mean8 = [0.0, 0.015, 0.044]  # 50% false
mean9 = [0.0, 0.008, 0.088]  # 100% miss
mean10 = [0.0, 0.016, 0.042] # 100% false


plt.figure(figsize=(20, 16))

N     = 3
ind   = np.arange(N)    # the x locations for the groups
width = 0.16       # the width of the bars: can also be len(x) sequence
b     = 'tab:blue'

pos1 = ind - 2*width - 0.02
pos2 = ind - width   - 0.01
pos3 = ind                  
pos4 = ind + width   + 0.01       
pos5 = ind + 2*width + 0.02 


p1 = plt.bar(pos1, mean1, width, color= b,  edgecolor='black', label='Miss')
p2 = plt.bar(pos1, mean2, width, color='r', edgecolor='black', bottom=mean1, label='False Alarm')

p3 = plt.bar(pos2, mean3, width, color= b,  edgecolor='black')
p4 = plt.bar(pos2, mean4, width, color='r', edgecolor='black', bottom=mean3)

p5 = plt.bar(pos3, mean5, width, color= b,  edgecolor='black')
p6 = plt.bar(pos3, mean6, width, color='r', edgecolor='black', bottom=mean5)

p7 = plt.bar(pos4, mean7, width, color= b,  edgecolor='black')
p8 = plt.bar(pos4, mean8, width, color='r', edgecolor='black', bottom=mean7)

p9 = plt.bar(pos5, mean9, width, color= b,  edgecolor='black')
p10 = plt.bar(pos5, mean10, width, color='r', edgecolor='black', bottom=mean9)

minor_pos = np.concatenate((pos1, pos2, pos3, pos4, pos5))
minor_lab = ['5%']*N + ['10%']*N + ['20%']*N + ['50%']*N + ['100%']*N
ax = plt.gca()
ax.set_xticks(minor_pos, minor=True)
ax.set_xticklabels(minor_lab, minor=True, fontsize=14, rotation=20)

plt.xticks(ind, ('1', '5','10'))
plt.xlabel('Varing # of Intruders')
ax.tick_params(axis='x', which='major', pad=35)

plt.legend()
plt.ylabel('Percentage')
plt.title('Miss & False Alarm by Percentage of Training Data and # of Intruders')
plt.savefig('varyIntruder-miss-false.png')
```
